```python
# import the necessary packages
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
# from numba import jit, cuda
```

```python
def mse(imageA, imageB):
	err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
	err /= float(imageA.shape[0] * imageA.shape[1])
	
	# return the MSE, the lower the error, the more "similar"
	return err
```

```python
dictionary = ['ANGER', 'CONTEMPT', 'DISGUST', 'FEAR', 'HAPPINESS',  'NEUTRAL', 'SADNESS', 'SURPRISE']
cwd = os.getcwd()

for emo in dictionary:
    print(emo)
    
    img_list = os.listdir(cwd + "/images/" + emo)

    # same scale
    scale_percent = 100

    for i in range(len(img_list)):
        if not os.path.exists("images/" + emo + "/" + img_list[i]):
            continue

        img1 = cv2.imread("images/" + emo + "/" + img_list[i])
        
        # set size
        width = int(img1.shape[1] * scale_percent / 100)
        height = int(img1.shape[0] * scale_percent / 100)
        dim = (width, height)
        
        # resize for comparation
        img1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)

        # convert the images to grayscale
        img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        
        for j in range(len(img_list)):
            # do not run same image
            if img_list[i] == img_list[j]:
                continue
            
            # do not run deleted image
            if not os.path.exists("images/" + emo + "/" + img_list[j]):
                continue

            img2 = cv2.imread("images/" + emo + "/" + img_list[j])

            # resize for comparation
            img2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)

            # convert the images to grayscale
            img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

            # compare the images
            m = mse(img1, img2)

            if m == 0:
                print(img_list[i], img_list[j])
                os.remove("images/" + emo + "/" + img_list[j])
```

```python
import csv
import pandas as pd

row = []
for emo in dictionary:
    for i in range(len(img_list)):
        path = "images/" + emo + "/" + img_list[i]

        if not os.path.exists("images/" + emo + "/" + img_list[i]):
            continue

        row.append([path, emo])

cwd = os.getcwd()
df = pd.read_csv(cwd + "/data/legend.csv")
df.to_csv(cwd + "/data_csv/all_data.csv")

f = open("all_data.csv", "a")
writer = csv.writer(f)
writer.writerow(row)
f.close()
```
